```python
import data_gradients
import torchvision
```

```python
from torchvision.datasets.mnist import FashionMNIST
from torchvision.transforms import Compose, ToTensor
```

```python
train = FashionMNIST(root='./data', train=True, download=True, transform=Compose([ToTensor()]))
valid = FashionMNIST(root='./data', train=False, download=True, transform=Compose([ToTensor()]))

len(train), len(valid)
```

```python
from torch.utils.data import DataLoader
from data_gradients.managers.classification_manager import ClassificationAnalysisManager

manager = ClassificationAnalysisManager(
    train_data=DataLoader(train),
    val_data=DataLoader(valid),
    report_title='Fashion MNIST',
    class_names=train.classes,
    n_image_channels=1,
)
```

```python
manager.run()
```
